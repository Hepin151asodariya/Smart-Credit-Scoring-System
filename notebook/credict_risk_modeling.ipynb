```python
features = ['Age', 'Sex', 'Job', 'Housing', 'Saving accounts', 'Checking account',
       'Credit amount', 'Duration']
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import joblib
```

```python
cat_cols = df_model.select_dtypes(include = "object").columns.drop("Risk")
```

```python
le_dict={}
```

```python
cat_cols
```

```python
# Instantiate OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fill NaN values in categorical columns with 'unknown'
df_model[cat_cols] = df_model[cat_cols].fillna("unknown")

# Fit and transform the categorical features
encoded_features = ohe.fit_transform(df_model[cat_cols])

# Get new column names
encoded_feature_names = ohe.get_feature_names_out(cat_cols)

# Create a DataFrame for the one-hot encoded features
df_encoded = pd.DataFrame(encoded_features, columns=encoded_feature_names, index=df_model.index)

# Drop original categorical columns from df_model
df_model = df_model.drop(columns=cat_cols)

# Concatenate df_model (now without original categorical features) and df_encoded
df_model = pd.concat([df_model, df_encoded], axis=1)

# Save the OneHotEncoder
joblib.dump(ohe, "onehot_encoder.joblib")
```

```python
le_target = LabelEncoder()
df_model[target] = le_target.fit_transform(df_model[target])
```

#Training models

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import GridSearchCV
```

```python
def train_model(model, param_grid,X_train,y_train,X_test,y_test):
    grid = GridSearchCV(model,param_grid,cv=5,scoring="accuracy",n_jobs=-1)
    grid.fit(X_train,y_train)
    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test)
    acc = accuracy_score(y_test,y_pred)

    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Check if the model has predict_proba, which is necessary for roc_auc_score
    if hasattr(best_model, "predict_proba"):
        y_pred_proba = best_model.predict_proba(X_test)[:, 1]
        roc_auc = roc_auc_score(y_test, y_pred_proba)
    else:
        roc_auc = None # Or handle appropriately if predict_proba is not available

    return best_model, acc, precision, recall, f1, roc_auc, grid.best_params_
```

#Decision Tree

```python
dt = DecisionTreeClassifier(random_state=1,class_weight="balanced")
dt_param_grid = {
    "max_depth":[3,5,7,10,None],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
}
```

```python
best_dt, acc_dt, precision_dt, recall_dt, f1_dt, roc_auc_dt, params_dt = train_model(dt,dt_param_grid,X_train,y_train,X_test,y_test)
```

```python
print("Decision Tree Accuracy",acc_dt)
print(f"Decision Tree Precision: {precision_dt:.4f}")
print(f"Decision Tree Recall: {recall_dt:.4f}")
print(f"Decision Tree F1-Score: {f1_dt:.4f}")
print(f"Decision Tree ROC-AUC: {roc_auc_dt:.4f}")
print("Best Parameters",params_dt)
```

#Random Forest

```python
rf = RandomForestClassifier(random_state=1,class_weight="balanced",n_jobs=-1)
rf_param_grid = {
    "n_estimators":[100,200],
    "max_depth":[3,5,7,10,None],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
}
```

```python
best_rf, acc_rf, precision_rf, recall_rf, f1_rf, roc_auc_rf, params_rf = train_model(rf,rf_param_grid,X_train,y_train,X_test,y_test)
```

```python
print("Random Forest Accuracy",acc_rf)
print(f"Random Forest Precision: {precision_rf:.4f}")
print(f"Random Forest Recall: {recall_rf:.4f}")
print(f"Random Forest F1-Score: {f1_rf:.4f}")
print(f"Random Forest ROC-AUC: {roc_auc_rf:.4f}")
print("Best Parameters",params_rf)
```

#XGBOOST

```python
xgb = XGBClassifier(random_state=1, scale_pos_weight = (y_train == 0).sum()/(y_train == 1).sum(),use_label_encoder=False,eval_metric = "logloss")
xgb_param_grid = {
  "n_estimators": [100, 200, 300],
    "max_depth": [3, 4, 5, 6],   # keep small
    "learning_rate": [0.01, 0.05, 0.1],  # remove 0.001
    "subsample": [0.7, 0.8, 0.9],
    "colsample_bytree": [0.7, 0.8, 0.9]
}
```

```python
best_xgb, acc_xgb, precision_xgb, recall_xgb, f1_xgb, roc_auc_xgb, params_xgb = train_model(xgb,xgb_param_grid,X_train,y_train,X_test,y_test)
```

```python
print("XGBOOST Accuracy",acc_xgb)
print(f"XGBOOST Precision: {precision_xgb:.4f}")
print(f"XGBOOST Recall: {recall_xgb:.4f}")
print(f"XGBOOST F1-Score: {f1_xgb:.4f}")
print(f"XGBOOST ROC-AUC: {roc_auc_xgb:.4f}")
print("Best Parameters",params_xgb)
```
